```python
def generarIntersecciones(calles, archivo, nfilas, ncol):
  n = len(calles)
  if n == nfilas + ncol:
    #Asignar numeros a las calles
    num = 0
    dicc_c = {}
    for calle in calles:
      dicc_c[calle] = num
      num += 1
    
    #Inicializamos el mapa que simule las intersecciones con puros -1  
    I = np.zeros([nfilas,ncol], int)
    I = I - 1

    #Actualizamos la matriz, poniendo los nodos de intersecciones
    with open(archivo) as f:
      for line in f:
        nodo, fila, columna = [elem for elem in line.split()]
        I[dicc_c[fila], dicc_c[columna] - nfilas] = int(nodo)

  return I
```

```python
I = generarIntersecciones(calles, "Intersecciones.txt", 42, 44)
print(I)
```

```python
def generarListaAdyacencia(I):
  G = []
  nfilas, ncol = np.shape(I)
  
  #Agregar las adyacencias de cada nodo
  for i in range(nfilas):
    for j in range(ncol):
      if I[i,j] > -1:
        aux = []
        if (i > 0) and (I[i-1, j] > -1):
          aux.append(I[i-1, j])
        if (j > 0) and (I[i, j-1] > -1):
          aux.append(I[i, j-1])
        if (j < ncol-1) and (I[i, j+1] > -1):
          aux.append(I[i, j+1])
        if (i < nfilas-1) and (I[i+1, j] > -1):
          aux.append(I[i+1, j])
        G.append(aux)

  #Guardar en un archivo
  cont = 0
  with open("Lista_Adyacencia_Generado.txt", 'w') as archivo:
    for lista in G:
      for num in lista:
        archivo.write(str(num) + " ")
      archivo.write("\n")

  return G
```
